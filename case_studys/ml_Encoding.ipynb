```python
tips['size']
```

```python
tips['size'].unique()
# if size < 2 == small group (s)
# is size 3,4 == medium group (m)
# if size 5,6 == large group (l)
```

```python
def order_data(x):
    if x > 4:
        return 'L'
    elif x > 2:
        return 'M'
    else :
        return "S"
```

```python
tips['size'].apply(order_data)
```

```python
tips['size'].apply(order_data).unique()
```

```python
tips["size_category"] = tips['size'].apply(order_data)
tips["size_category"]
```

```python
tips
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder().set_output(transform='pandas')
oe
```

```python
oe.fit_transform(tips[['size_category']])
```

```python
# to maintain a order 
order = [ 'S','M','L']
```

```python
oe1 = OrdinalEncoder( categories=[order]).set_output(transform='pandas')
oe1
```

```python
oe1.fit_transform(tips[['size_category']])
```

```python
## Encoding 
ohe = OneHotEncoder(sparse_output=False , drop='first')
oe = OrdinalEncoder(categories=[order])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer([('OHE', ohe, ['sex','smoker', 'day' , 'time']),
                        ('OE', oe,['size_category'])],
                        remainder='passthrough').set_output(transform='pandas')
```

```python
ct
```

```python
ct.fit_transform(tips)
```

```python
ct.fit_transform(tips[:])
```
