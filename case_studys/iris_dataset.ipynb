# IQR method 
- Q1=25 , Q3=75
- IQR = Q3 - Q1
- 1.5*IQR 
- Q1 - 1.5IQR = ll lower limmit
- Q3 + 1.5IQR = ul upper limmit 
- compair minimum value with lower limit , if min < ll = outliers
- compair maximum value with upper limit , if max > ul = outliers 
- values between lower limit and upper limit are non_outlier values [ ll < values >ul]
- values which are less them lower limit and greater them upper limit are outliers [outliers>ll, ul<outliers ]
