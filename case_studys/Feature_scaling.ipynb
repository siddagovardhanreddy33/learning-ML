```python
from sklearn.preprocessing import RobustScaler
```

```python
R_scalar = RobustScaler().set_output(transform="pandas")
R_scalar
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer(
    [('Robust_scalar', R_scalar , ['total_bill','tip'])]
).set_output(transform='pandas')

ct
```

```python
ct.fit_transform(tips)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False, drop='first')
ohe
```

```python
ct2 = ColumnTransformer(
    [
        ('R_scale' ,R_scalar, ['total_bill' , 'tip']),
        ('Nominal' ,ohe, ['sex' ,'smoker' ,'day' ,'time'])
        
    ]
).set_output(transform='pandas')

ct2
```

```python
ct2 = ColumnTransformer(
    [
        ('Scale', R_scalar, ['total_bill', 'tip']),
        ('Nominal', ohe, ['sex', 'smoker', 'day', 'time'])
    ]
).set_output(transform='pandas')

ct2
```

```python
ct2.fit_transform(tips)
```

```python
ct2 = ColumnTransformer(
    [
        ('Scale', R_scalar, ['total_bill', 'tip']),
        ('Nominal', ohe, ['sex', 'smoker', 'day', 'time'])
    ], remainder='passthrough'   # gives all the remaining columns in output
).set_output(transform='pandas')

ct2
```

```python
ct2.fit_transform(tips)
```

```python
ct2 = ColumnTransformer(
    [
        ('Scale', R_scalar, ['total_bill', 'tip']),
        ('Nominal', ohe, ['sex', 'smoker', 'day', 'time'])
    ], remainder='passthrough',   # gives all the remaining columns in output
    verbose_feature_names_out=False   # no prefixs are added to the column names

).set_output(transform='pandas')

ct2
```

```python
ct2.fit_transform(tips)
```

```python
def order_data(x):
    if x > 4:
        return 'L'
    elif x > 2:
        return 'M'
    else :
        return 'S'
```

```python
tips['size'] = tips['size'].apply(order_data)
tips.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
order = ['S','M','L']  # ascending order

oe = OrdinalEncoder(categories=[order])

oe
```

```python
ct3 = ColumnTransformer(
    [
        ('R_scalar',R_scalar,['total_bill','tip']),
        ('Nominal',ohe,['sex', 'smoker', 'day', 'time']),
        ('Ordinal',oe , ['size']),
    ],
    remainder='passthrough',
    verbose_feature_names_out=False
).set_output(transform='pandas')

ct3
```

```python
ct3.fit_transform(tips)
```
