```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split

from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler

from sklearn.pipeline import Pipeline , make_pipeline

from sklearn.feature_selection import SelectKBest , chi2

from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('train.csv')
df
```

## planing

```python
df.drop(columns=['PassengerId', 'Name' , 'Ticket' , 'Cabin'] , inplace=True)
df
```

```python
# step 1 ---> train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train , y_test = train_test_split(df.drop(columns=['Survived']),
                                                     df['Survived'],
                                                     test_size=0.2,
                                                     random_state=42)
```

```python
# imputation transformation
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer


trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')
```

```python
trf1
```

```python
# one hot encoder

trf2 = ColumnTransformer([
    ('OneHotEncoder', OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])
],remainder='passthrough')
```

```python
trf2
```

```python
## scaling
from sklearn.preprocessing import MinMaxScaler

trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
trf3
```

```python
# Feature selection
from sklearn.feature_selection import SelectKBest , chi2
trf4 = SelectKBest(score_func=chi2 , k=8)
```

```python
trf4
```

```python
# train the model 
from sklearn.tree import DecisionTreeClassifier

trf5 = DecisionTreeClassifier()
```

```python
trf5
```

## Creating pipeline

```python
from sklearn.pipeline import Pipeline , make_pipeline
```

### pipeline vs make_pipeline
- pipeline requires naming of steps 
- make_pipeline doesn't need any naming of steps
- note: (same applies to ColumnTransformer vs make_Column_transformer)

```python
# 1 make_pipeline

pipe = make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
pipe
```

```python
# 2 pipeline

pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5),
])
```

```python
pipe
```

```python
# train

pipe.fit(X_train,y_train)
```

## Explore the pipeline

```python
pipe.named_steps
```

```python
pipe.named_steps['trf1'].transformers_
```

```python
pipe.named_steps['trf1'].transformers_[0][1].statistics_
```

```python
pipe.named_steps['trf1'].transformers_[1][1].statistics_
```

```python
# Display pipeline

from sklearn import set_config
set_config(display='diagram')
```

```python
# predict

y_pred = pipe.predict(X_test)
y_pred
```

```python
# 
from sklearn.metrics import accuracy_score

accuracy_score(y_test , y_pred)
```

## cross validation using Pipeline

```python
# cross validation using cross_val_score

from sklearn.model_selection import cross_val_score

cross_val_score(pipe , X_train , y_train , cv=5 , scoring='accuracy').mean()
```

## GridSearch using Pipeline
- Hyperparameter tuning

```python
# gridsearchcv

params = {
    'trf5__max_depth':[1,2,3,4,5,None]
}
```

```python
from sklearn.model_selection import GridSearchCV

Grid = GridSearchCV(pipe , params , cv=5 , scoring='accuracy')
Grid
```

```python
Grid.fit(X_train , y_train)
```

```python
Grid.best_score_
```
