```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df =  pd.read_csv('train.csv')

df
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'], inplace=True)
```

```python
df
```

```python
# steo 1 --> train test split

from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(df.drop(columns=['Survived']),
                                                       df['Survived'],
                                                       test_size=0.2,
                                                       random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# Applying missing values using simple imputatuin

from sklearn.impute import  SimpleImputer
```

```python
df.isnull().sum()
```

```python
si_age = SimpleImputer()
si_age
```

```python
si_embarked = SimpleImputer(strategy='most_frequent')
si_embarked
```

```python
X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])
```

```python
X_test_age = si_age.transform(X_test[['Age']])
X_test_embarked = si_embarked.transform(X_test[['Embarked']])
```

```python
# OneHotEncoding  sex and Embarked
from sklearn.preprocessing import OneHotEncoder

ohe_sex = OneHotEncoder(sparse_output=False , handle_unknown='ignore')
ohe_sex
```

```python
ohe_embarked =  OneHotEncoder(sparse_output=False , handle_unknown='ignore')
ohe_embarked
```

```python
# trained data
X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)
```

```python
X_train_sex
```

```python
X_train_embarked
```

```python
# Test data
X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X_test_embarked)
```

```python
X_test_sex
```

```python
X_test_embarked
```

### concatination

```python
# droping columns
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])

X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
# concatination

X_train_transformed = np.concatenate((X_train_rem , X_train_age , X_train_embarked , X_train_sex),axis=1)
X_train_transformed
```

```python
X_test_transformed = np.concatenate((X_test_rem , X_test_age , X_test_embarked , X_test_sex), axis=1)

X_test_transformed
```

```python
# model prediction 

clf = DecisionTreeClassifier()
clf
```

```python
clf.fit(X_train_transformed,y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
y_pred
```

```python
# accuresy testing 

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
